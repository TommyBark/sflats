# Example of use

Firstly, we import sflats and load the dataset.

```python
import sflats
```

```python
A = sflats.Flats()
```

```python
df = A.load_dataset()
```

We can check our data.

```python
df.head()
```

```python
A.load_dataset(remove_labels=True,latest_only=True).head()
```

### Plot district map , color by average price per m^2, using all data

We can now plot district maps. District maps may be helpful to identify differences in average district characteristics and can thus give aid during a house search. To enhance the visualization we can color the districts according to these average characteristics. Firstly, we can do so for average flat price per meter square (Ppm2).

```python
A.district_map(color_by='Ppm2')
```

### Plot district map , color by average price , using only latest data

We can also color districts according to average flat price. (Another possible option is to color districts according to average area.)

```python
A.district_map(color_by='Price',latest_only=False)
```

### Plot all latest flats on map, markers are colored by price per m^2

One can also view all of the latest flats on a map, together with their respective characteristics. This can help one view available flats and potentially choose a specific one. Additionally, there are also sliders so that one can specify and view the flats that are of specific interest. Moreover, the urls of the flats are also provided, enabling one to easily view the websites of the flats.   

Sliders need to be initially changed for the map to update.

```python
A.map()
```

### Plot quantile plots, count plot and pie chart for all flats

Quantile plots for various features, such as flat area, price or price per square meter in time are provided. One can also view a plot showing the number of available flats over time as well as a pie chart for proportions of different flat types. These plots can give information about the housing situation in Prague and any trends or changes over time. They can thus help the user in the search for flats. 

Lowest line in the quantile plots show 10% quantile, Highest line 90% quantile

```python
A.plots()
```

### Plot quantile plots, count plot and pie chart only for flats with 1 room 

One can also choose to display these plots only for certain flat types that are of interest. For example, we can view them only for 1+kk and 1+1 flats.

```python
A.plots(flat_type=['1+kk','1+1'])
```
